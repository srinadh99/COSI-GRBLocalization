```python
import numpy as np
from tqdm import tqdm
import time
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# ===================================================================
#                      1. HELPER FUNCTIONS
# ===================================================================

def transform_event_features(event_sets, angular_feature_indices):
    """Applies sin/cos transformation to specified angular features."""
    transformed_event_sets = []
    if not angular_feature_indices: return [events.copy() for events in event_sets]
    
    num_non_angular = len([i for i in range(event_sets[0].shape[1]) if i not in angular_feature_indices])
    output_dim = num_non_angular + 2 * len(angular_feature_indices)

    for events in event_sets:
        if events.shape[0] == 0:
            transformed_event_sets.append(np.empty((0, output_dim), dtype=np.float32))
            continue
        non_angular_indices = [i for i in range(events.shape[1]) if i not in angular_feature_indices]
        new_feature_parts = [events[:, non_angular_indices]]
        for idx in angular_feature_indices:
            angles_rad = np.deg2rad(events[:, idx])
            new_feature_parts.extend([np.cos(angles_rad)[:, np.newaxis], np.sin(angles_rad)[:, np.newaxis]])
        transformed_events = np.concatenate(new_feature_parts, axis=1).astype(np.float32)
        transformed_event_sets.append(transformed_events)
    return transformed_event_sets

def convert_to_cartesian(coords_deg):
    """Converts (longitude, latitude) degrees to a 3D Cartesian unit vector."""
    lon_deg, lat_deg = coords_deg[:, 0], coords_deg[:, 1]
    lon_rad, lat_rad = np.deg2rad(lon_deg), np.deg2rad(lat_deg)
    x = np.cos(lon_rad) * np.cos(lat_rad)
    y = np.sin(lon_rad) * np.cos(lat_rad)
    z = np.sin(lat_rad)
    return np.stack([x, y, z], axis=1).astype(np.float32)

def convert_cartesian_to_lon_lat(cartesian_coords):
    """Converts 3D Cartesian vectors to (longitude, latitude) degrees."""
    # Ensure input is a 2D array
    if cartesian_coords.ndim == 1:
        cartesian_coords = cartesian_coords[np.newaxis, :]
        
    norms = np.linalg.norm(cartesian_coords, axis=1, keepdims=True)
    norms[norms == 0] = 1 # Avoid division by zero
    
    cartesian_normalized = cartesian_coords / norms
    x, y, z = cartesian_normalized[:, 0], cartesian_normalized[:, 1], cartesian_normalized[:, 2]
    
    lat_rad = np.arcsin(np.clip(z, -1.0, 1.0))
    lon_rad = np.arctan2(y, x)
    
    lat_deg = np.rad2deg(lat_rad)
    lon_deg = np.rad2deg(lon_rad)
    lon_deg = (lon_deg + 360) % 360
    
    return np.stack([lon_deg, lat_deg], axis=1)

def calculate_angular_error(y_true_cartesian, y_pred_cartesian):
    """Calculates the angular separation in degrees."""
    pred_norms = np.linalg.norm(y_pred_cartesian, axis=1, keepdims=True)
    pred_norms[pred_norms == 0] = 1
    y_pred_normalized = y_pred_cartesian / pred_norms
    
    dot_product = np.clip(np.sum(y_true_cartesian * y_pred_normalized, axis=1), -1.0, 1.0)
    angle_rad = np.arccos(dot_product)
    return np.rad2deg(angle_rad)

# ===================================================================
# ### NEW: FUNCTION FOR UNCERTAINTY PROPAGATION ###
# ===================================================================
def propagate_cartesian_to_spherical_uncertainty_mc(mu_cart, var_cart, num_samples=5000):
    """
    Propagates uncertainty from Cartesian to spherical coordinates using Monte Carlo.

    Args:
        mu_cart (np.ndarray): Array of mean Cartesian vectors, shape (N, 3).
        var_cart (np.ndarray): Array of Cartesian variances (diagonal), shape (N, 3).
        num_samples (int): Number of MC samples to draw for each point.

    Returns:
        np.ndarray: Array of standard deviations in (lon, lat), shape (N, 2).
    """
    num_predictions = mu_cart.shape[0]
    std_lb = np.zeros((num_predictions, 2))

    print(f"Propagating uncertainty for {num_predictions} predictions using {num_samples} MC samples...")
    for i in tqdm(range(num_predictions)):
        # If variance is extremely small, uncertainty is zero.
        if np.all(var_cart[i] < 1e-9):
            std_lb[i] = [0, 0]
            continue
            
        # 1. Draw samples from the 3D Gaussian distribution
        cov_matrix = np.diag(var_cart[i])
        samples_cart = np.random.multivariate_normal(mu_cart[i], cov_matrix, size=num_samples)

        # 2. Convert all samples to spherical coordinates
        samples_lb = convert_cartesian_to_lon_lat(samples_cart)
        
        # 3. Calculate standard deviation for latitude (b) - straightforward
        std_lat = np.std(samples_lb[:, 1])

        # 4. Calculate standard deviation for longitude (l) - using circular statistics
        lon_rad = np.deg2rad(samples_lb[:, 0])
        # Convert angles to vectors on a unit circle
        x_comp = np.cos(lon_rad)
        y_comp = np.sin(lon_rad)
        # Average the vectors
        mean_x = np.mean(x_comp)
        mean_y = np.mean(y_comp)
        # Length of the mean vector (R)
        r = np.sqrt(mean_x**2 + mean_y**2)
        # Circular standard deviation formula
        circular_std_rad = np.sqrt(-2 * np.log(r))
        std_lon = np.rad2deg(circular_std_rad)
        
        std_lb[i] = [std_lon, std_lat]
        
    return std_lb

# ===================================================================
#           2. PROBABILISTIC DEEP SETS MODEL (Unchanged)
# ===================================================================
class ProbabilisticDeepSet(nn.Module):
    def __init__(self, input_dim, phi_hidden_dims, latent_dim, rho_hidden_dims):
        super(ProbabilisticDeepSet, self).__init__()
        phi_layers = [nn.Linear(input_dim, phi_hidden_dims[0]), nn.ReLU()]
        for i in range(len(phi_hidden_dims) - 1):
            phi_layers.extend([nn.Linear(phi_hidden_dims[i], phi_hidden_dims[i+1]), nn.ReLU()])
        phi_layers.append(nn.Linear(phi_hidden_dims[-1], latent_dim))
        self.phi = nn.Sequential(*phi_layers)
        
        rho_layers = [nn.Linear(latent_dim, rho_hidden_dims[0]), nn.ReLU()]
        for i in range(len(rho_hidden_dims) - 1):
            rho_layers.extend([nn.Linear(rho_hidden_dims[i], rho_hidden_dims[i+1]), nn.ReLU()])
        rho_layers.append(nn.Linear(rho_hidden_dims[-1], 6))
        self.rho = nn.Sequential(*rho_layers)

    def forward(self, x_list):
        reps = []
        for events in x_list:
            if events.numel() == 0:
                reps.append(torch.zeros(self.rho[0].in_features, device=self.rho[0].weight.device))
            else:
                reps.append(torch.sum(self.phi(events), dim=0))
        rho_output = self.rho(torch.stack(reps))
        mu, log_var = torch.chunk(rho_output, 2, dim=1)
        return mu, log_var

# ===================================================================
#           3. PROBABILISTIC LOSS FUNCTION (NLL) (Unchanged)
# ===================================================================
def nll_loss(y_true, mu, log_var):
    mu_normalized = mu / torch.linalg.norm(mu, dim=1, keepdim=True)
    var = torch.exp(log_var)
    log_det_sigma = torch.sum(log_var, dim=1)
    inv_sigma_term = torch.sum((y_true - mu_normalized)**2 / var, dim=1)
    k = y_true.shape[1]
    nll = 0.5 * (log_det_sigma + inv_sigma_term + k * np.log(2 * np.pi))
    return torch.mean(nll)

# ===================================================================
#              4. CUSTOM DATASET AND DATALOADER (Unchanged)
# ===================================================================
class ComptonDataset(Dataset):
    def __init__(self, x_events, y_targets, scaler=None):
        self.x_events = x_events
        self.y_targets = y_targets
        self.scaler = scaler

    def __len__(self): return len(self.x_events)
    def __getitem__(self, idx):
        events, target = self.x_events[idx], self.y_targets[idx]
        if self.scaler and events.shape[0] > 0:
            events = self.scaler.transform(events)
        return torch.FloatTensor(events), torch.FloatTensor(target)

def collate_fn(batch):
    events_list = [item[0] for item in batch]
    targets = torch.stack([item[1] for item in batch])
    return events_list, targets

# ===================================================================
#                 ### MODIFIED: 5. MAIN SCRIPT ###
# ===================================================================
if __name__ == '__main__':
    # --- Configuration ---
    ANGULAR_FEATURE_INDICES = [1, 2, 4] # l1, b1, l2
    PHI_HIDDEN_DIMS = [128, 256, 512]
    LATENT_DIM = 1024
    RHO_HIDDEN_DIMS = [512, 256, 128, 64]
    
    LEARNING_RATE = 1e-4
    BATCH_SIZE = 64
    EPOCHS = 100 # Set back to a reasonable number for full training
    DEVICE = "cuda:0" if torch.cuda.is_available() else "cpu"
    
    # --- 1. Load and Preprocess Data ---
    print("--- 1: Loading and Preprocessing Data ---")
    x_events_raw = np.load("./All_Events_SignalBkg5_Deg.npy", allow_pickle=True)
    y_coords = np.load("./Ground_Truths.npy", allow_pickle=True)
    y_lon_lat_deg_true = np.array(y_coords[:, 1:3], dtype=np.float32)
    y_lon_lat_deg_lmmlem = np.array(y_coords[:, 3:5], dtype=np.float32)
    x_events_transformed = transform_event_features(x_events_raw, ANGULAR_FEATURE_INDICES)
    y_cartesian_true = convert_to_cartesian(y_lon_lat_deg_true)
    
    # --- 2. Create Splits and Fit Scaler ---
    print("\n--- 2: Creating Data Splits and Scaler ---")
    splits = train_test_split(x_events_raw, x_events_transformed, y_cartesian_true, y_lon_lat_deg_true,
                              y_lon_lat_deg_lmmlem, test_size=0.2, random_state=42)
    x_raw_train, x_raw_test, x_train, x_test, y_train, y_test, y_train_lb, y_test_lb, y_lmmlem_train, y_lmmlem_test = splits
    all_train_events = np.concatenate([e for e in x_train if e.shape[0] > 0])
    scaler = StandardScaler().fit(all_train_events)
    
    # --- 3. Create Datasets and DataLoaders ---
    print("\n--- 3: Setting up Datasets and DataLoaders ---")
    train_dataset = ComptonDataset(x_train, y_train, scaler)
    test_dataset = ComptonDataset(x_test, y_test, scaler)
    train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True, collate_fn=collate_fn)
    test_loader = DataLoader(test_dataset, batch_size=BATCH_SIZE, shuffle=False, collate_fn=collate_fn)
    
    # --- 4. Initialize Model, Loss, and Optimizer ---
    print("\n--- 4: Initializing Model and Training Components ---")
    model = ProbabilisticDeepSet(all_train_events.shape[1], PHI_HIDDEN_DIMS, LATENT_DIM, RHO_HIDDEN_DIMS).to(DEVICE)
    optimizer = optim.Adam(model.parameters(), lr=LEARNING_RATE)
    print(f"Using device: {DEVICE}\n{model}")

    # --- 5. Training Loop ---
    print("\n--- 5: Starting Training ---")
    for epoch in range(EPOCHS):
        model.train()
        running_loss = 0.0
        for events, targets in tqdm(train_loader, desc=f"Epoch {epoch+1}/{EPOCHS}"):
            events = [e.to(DEVICE) for e in events]
            targets = targets.to(DEVICE)
            optimizer.zero_grad()
            pred_mu, pred_log_var = model(events)
            loss = nll_loss(targets, pred_mu, pred_log_var)
            loss.backward()
            optimizer.step()
            running_loss += loss.item()
        print(f"Epoch {epoch+1}/{EPOCHS}, Loss: {running_loss / len(train_loader):.6f}")

    # --- 6. Evaluation with LMMLEM Comparison --- ### MODIFIED ###
    print("\n--- 6: Evaluating Models on Test Set ---")
    model.eval()
    all_mu, all_log_var, all_targets = [], [], []
    with torch.no_grad():
        for events, targets in tqdm(test_loader, desc="Evaluating Deep Set"):
            mu, log_var = model([e.to(DEVICE) for e in events])
            all_mu.append(mu.cpu().numpy())
            all_log_var.append(log_var.cpu().numpy())
            all_targets.append(targets.cpu().numpy())
            
    y_pred_cart_deepset = np.concatenate(all_mu)
    y_pred_log_var_deepset = np.concatenate(all_log_var)
    y_true_cart_eval = np.concatenate(all_targets)
    
    angular_errors_deepset = calculate_angular_error(y_true_cart_eval, y_pred_cart_deepset)
    angular_errors_lmmlem = calculate_angular_error(y_true_cart_eval, convert_to_cartesian(y_lmmlem_test))
    
    # ### NEW: Propagate uncertainty to spherical coordinates ###
    y_pred_var_deepset = np.exp(y_pred_log_var_deepset)
    y_pred_std_lb = propagate_cartesian_to_spherical_uncertainty_mc(y_pred_cart_deepset, y_pred_var_deepset)
    
    print(f"\n{'='*15} Final Model Comparison {'='*15}")
    # ... (Performance summary prints are the same) ...
    
    # --- Detailed Sample Predictions --- ### MODIFIED ###
    y_pred_lb_deepset = convert_cartesian_to_lon_lat(y_pred_cart_deepset)
    print("\n--- Sample Predictions (l, b in degrees) with Spherical Uncertainty ---")
    for i in range(min(5, len(x_raw_test))):
        true_l, true_b = y_test_lb[i]
        lmmlem_l, lmmlem_b = y_lmmlem_test[i]
        deepset_l, deepset_b = y_pred_lb_deepset[i]
        std_l, std_b = y_pred_std_lb[i]
        
        print(f"\nSample {i+1} (Events: {len(x_raw_test[i])}):")
        print(f"  - Ground Truth:     ({true_l:7.2f}, {true_b:6.2f})")
        print(f"  - LMMLEM Pred:      ({lmmlem_l:7.2f}, {lmmlem_b:6.2f}) -> Error: {angular_errors_lmmlem[i]:.2f}°")
        print(f"  - Deep Sets Pred:   ({deepset_l:7.2f}, {deepset_b:6.2f}) -> Error: {angular_errors_deepset[i]:.2f}°")
        print(f"  - Deep Sets Uncert (σ_l, σ_b in deg): ({std_l:6.2f}, {std_b:6.2f})")
    
    # ===================================================================
    #              ### MODIFIED: 7. VISUALIZATION OF RESULTS ###
    # ===================================================================
    print("\n--- 7: Visualizing Results ---")
    test_event_counts = np.array([len(e) for e in x_raw_test])

    # Plot 1: Angular Error vs. Number of Events (Unchanged)
    plt.figure(figsize=(12, 7))
    plt.scatter(test_event_counts, angular_errors_lmmlem, alpha=0.5, s=20, label='LMMLEM Error', c='orangered')
    plt.scatter(test_event_counts, angular_errors_deepset, alpha=0.5, s=20, label='Deep Sets Error', c='dodgerblue')
    plt.title('Model Angular Error vs. Number of Events', fontsize=16)
    plt.xlabel('Number of Events in Source', fontsize=12)
    plt.ylabel('Angular Error (degrees)', fontsize=12)
    plt.xscale('log')
    plt.legend(fontsize=12)
    plt.grid(True, which='both', linestyle='--', linewidth=0.5)
    plt.tight_layout()
    plt.savefig("error_vs_events_comparison_100.png")
    print("\nGenerated plot 'error_vs_events_comparison_100.png'")
    plt.show()

    # ### MODIFIED Plot 2: Model Angular Uncertainty vs. Number of Events ###
    # Combine σ_l and σ_b into a single metric for total angular uncertainty.
    total_angular_std = np.sqrt(y_pred_std_lb[:, 0]**2 + y_pred_std_lb[:, 1]**2)
    
    plt.figure(figsize=(12, 7))
    scatter = plt.scatter(test_event_counts, total_angular_std, c=angular_errors_deepset, 
                          cmap='plasma', alpha=0.7, s=30)
    plt.title('Model Angular Uncertainty vs. Number of Events', fontsize=16)
    plt.xlabel('Number of Events in Source', fontsize=12)
    plt.ylabel('Predicted Total Angular Uncertainty (σ_angular in degrees)', fontsize=12)
    plt.xscale('log')
    plt.yscale('log')
    cbar = plt.colorbar(scatter)
    cbar.set_label('Actual Angular Error (degrees)')
    plt.grid(True, which='both', linestyle='--', linewidth=0.5)
    plt.tight_layout()
    plt.savefig("angular_uncertainty_vs_events_100.png")
    print("Generated plot 'angular_uncertainty_vs_events_100.png'")
    plt.show()
```

```python
# ... (All previous code, up to the end of the evaluation step) ...
# The following variables should have been computed right before this block:
# y_pred_cart_deepset, y_pred_var_deepset, y_true_cart_eval,
# angular_errors_deepset, angular_errors_lmmlem, y_pred_std_lb, y_test_lb


def plot_distribution_map(sample_index, y_true_all_lb, y_pred_mean_all_lb, 
                          mu_cart_all, var_cart_all,
                          num_mc_samples=10000, nside=32, fwhm_deg=1.5):
    """
    Generates and saves a Healpy sky map showing the predicted probability
    distribution for a single test sample.

    Args:
        sample_index (int): The index of the sample to plot from the test set.
        y_true_all_lb (np.ndarray): Array of all ground truth [lon, lat] pairs.
        y_pred_mean_all_lb (np.ndarray): Array of all predicted mean [lon, lat] pairs.
        mu_cart_all (np.ndarray): Array of all predicted Cartesian means.
        var_cart_all (np.ndarray): Array of all predicted Cartesian variances.
        num_mc_samples (int): Number of Monte Carlo samples to draw.
        nside (int): The NSIDE resolution for the Healpy map.
        fwhm_deg (float): The FWHM of the Gaussian smoothing kernel in degrees.
    """
    print(f"\n--- Generating distribution map for test sample {sample_index} ---")
    
    # --- 1. Select data for the chosen sample ---
    truth_lb = y_true_all_lb[sample_index]
    pred_mean_lb = y_pred_mean_all_lb[sample_index]
    mu_cart_pred = mu_cart_all[sample_index]
    var_cart_pred = var_cart_all[sample_index]

    # --- 2. Monte Carlo Simulation ---
    print(f"Generating {num_mc_samples} samples from the predicted distribution...")
    covariance_matrix = np.diag(var_cart_pred)
    # Generate samples from the 3D Gaussian distribution
    samples_cart = np.random.multivariate_normal(mu_cart_pred, covariance_matrix, size=num_mc_samples)
    # Project samples onto the sphere and convert to lon/lat
    samples_lb = convert_cartesian_to_lon_lat(samples_cart)

    # --- 3. Create Probability Density Map ---
    print("Binning samples into a high-resolution map...")
    npix_prob = hp.nside2npix(nside)
    prob_map = np.zeros(npix_prob)
    pix_indices = hp.ang2pix(nside, samples_lb[:, 0], samples_lb[:, 1], lonlat=True)
    counts = np.bincount(pix_indices, minlength=npix_prob)
    prob_map += counts

    # --- 4. Smooth the map ---
    print(f"Smoothing the map with a {fwhm_deg}° FWHM kernel...")
    prob_map_smoothed = hp.smoothing(prob_map, fwhm=np.radians(fwhm_deg))

    # --- 5. Create the Healpy Plot ---
    print("Creating final sky map...")
    hp.mollview(
        prob_map_smoothed,
        coord="G",
        title="",  # Title will be set manually for better positioning
        unit="Probability Density",
        cmap='Blues',
        min=0,
    )
    hp.graticule(coord="G", color='dimgray', alpha=0.7, dpar=30, dmer=30)

    # --- 6. Plot Ground Truth and Predicted Mean ---
    hp.projscatter(truth_lb[0], truth_lb[1],
                   lonlat=True, color="lime", s=150, marker="*", coord="G",
                   edgecolor='black', linewidth=1.5, label='Ground Truth')
    hp.projscatter(pred_mean_lb[0], pred_mean_lb[1],
                   lonlat=True, color="cyan", s=120, marker="P", coord="G",
                   edgecolor='black', linewidth=1.5, label='Deep Sets (Mean)')

    # --- 7. Final Touches ---
    fig = plt.gcf()
    ax = plt.gca()
    from matplotlib.lines import Line2D
    legend_elements = [
        Line2D([0], [0], marker='*', color='w', label='Ground Truth', markerfacecolor='lime', markeredgecolor='k', markersize=15),
        Line2D([0], [0], marker='P', color='w', label='Deep Sets (Mean)', markerfacecolor='cyan', markeredgecolor='k', markersize=12),
    ]
    ax.legend(handles=legend_elements, loc="upper right", fontsize=12, frameon=True)
    ax.set_title(f"Predicted Source Distribution (Sample {sample_index})", fontsize=16, y=1.05)

    # --- 8. Save and Show ---
    save_filename = f"distribution_map_sample_{sample_index}.png"
    plt.savefig(save_filename, dpi=200, bbox_inches="tight")
    print(f"Plot saved as '{save_filename}'")
    plt.show()


# ===================================================================
#              8. FINAL SUMMARY AND VISUALIZATIONS
# ===================================================================

# --- First, run the summary and overview plots as before ---
# (This code is the same as the previous response)

# --- Performance summary prints ---
print("\n\n" + "="*25 + " FINAL RESULTS SUMMARY " + "="*25)
# ... (code for printing the summary table) ...

# --- Detailed Sample Predictions ---
# ... (code for printing the text-based sample predictions) ...

# --- VISUALIZATION OF AGGREGATE RESULTS ---
# ... (code for the 'error_vs_events' and 'uncertainty_vs_events' scatter plots) ...


# ===================================================================
#      9. DETAILED VISUALIZATION OF INDIVIDUAL DISTRIBUTIONS
# ===================================================================

# Convert Deep Set cartesian predictions to spherical (lon/lat) for plotting
y_pred_lb_deepset = convert_cartesian_to_lon_lat(y_pred_cart_deepset)

# You can now easily plot the distribution for any sample.
# Just change the 'sample_index' to view a different result.

# Example 1: Plot a sample (e.g., the 6th sample in the test set)
plot_distribution_map(
    sample_index=1, 
    y_true_all_lb=y_test_lb,
    y_pred_mean_all_lb=y_pred_lb_deepset,
    mu_cart_all=y_pred_cart_deepset,
    var_cart_all=y_pred_var_deepset
)

# Example 2: Plot another sample (e.g., the 11th sample)
plot_distribution_map(
    sample_index=234, 
    y_true_all_lb=y_test_lb,
    y_pred_mean_all_lb=y_pred_lb_deepset,
    mu_cart_all=y_pred_cart_deepset,
    var_cart_all=y_pred_var_deepset,
    fwhm_deg=2.0 # You can also change parameters per call
)

# Example 3: Plot another sample (e.g., the 11th sample)
plot_distribution_map(
    sample_index=489, 
    y_true_all_lb=y_test_lb,
    y_pred_mean_all_lb=y_pred_lb_deepset,
    mu_cart_all=y_pred_cart_deepset,
    var_cart_all=y_pred_var_deepset,
    fwhm_deg=2.0 # You can also change parameters per call
)

# Example 4: Plot another sample (e.g., the 11th sample)
plot_distribution_map(
    sample_index=769, 
    y_true_all_lb=y_test_lb,
    y_pred_mean_all_lb=y_pred_lb_deepset,
    mu_cart_all=y_pred_cart_deepset,
    var_cart_all=y_pred_var_deepset,
    fwhm_deg=2.0 # You can also change parameters per call
)

# Example 5: Plot another sample (e.g., the 11th sample)
plot_distribution_map(
    sample_index=976, 
    y_true_all_lb=y_test_lb,
    y_pred_mean_all_lb=y_pred_lb_deepset,
    mu_cart_all=y_pred_cart_deepset,
    var_cart_all=y_pred_var_deepset,
    fwhm_deg=2.0 # You can also change parameters per call
)
```
